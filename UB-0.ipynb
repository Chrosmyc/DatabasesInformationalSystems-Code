<h4> Die folgende Beschreibung skizziert die Anforderungen für ein Datenbanksystem, das Informationen für ein Krankenhaus-Management-System enthält.</h4>

1. Es werden Informationen zu jedem `Patient` gespeichert. Hierzu gehören `Name`, eine oder mehrere `Telefonnummer`n, eine `Adresse` (bestehend aus `Strasse`, `PLZ` und `Stadt`) und eine eindeutige `Krankenversicherungsnummer`.
2. Ein `Arzt` kann von mehreren Patienten besucht werden. Ein Patient kann mehrere Ärzte besuchen. Ein Arzt kann unter anderem entweder `Assistenzarzt`, `Oberarzt` oder `Chefarzt` sein.
3. Ein Patient kann ein oder mehrere `Medikament`e erhalten für eine oder mehrere `Behandlung`en.
4. Jedes Medikament hat eine eindeutige `Pharmazentralnummer` (PZN).
5. Jede Behandlung hat eine `Beschreibung` und eindeutige `ID`.
6. Ein Arzt verschreibt ein oder mehrere Medikamente. Ein Medikament kann von einem oder mehreren Ärzten verschrieben werden.
7. Ein Mitarbeiter ist entweder ein `Arzt` oder ein `Krankenpfleger`. Andere Rollen werden nicht erwartet. Zu jedem Mitarbeiter werden `Name`, eine eindeutige `Personalnummer` und eine oder mehrere `Telefonnummern` gespeichert. Ein Mitarbeiter wird von einem Mitarbeiter beaufsichtigt, ein Mitarbeiter beaufsichtigt mehrere Mitarbeiter.
8. Ein Krankenpfleger arbeitet auf genau einer `Station`. In einer Station arbeiten mehrere Krankenpfleger.
9. Eine Station hat eine eindeutige `Bezeichnung`. Eine Station enthält mehrere `Zimmer`. Die `Zimmernummer` ist innerhalb einer Station eindeutig. Wird eine Station entfernt, wird das dazugehörige Zimmer auch entfernt. 
10. Jedem Patienten ist an einem `Datum` genau ein Zimmer zugeteilt. In jedem Zimmer ist immer genau ein Patient.

```python
g = ER()
diags = []

# 1. Es werden Informationen zu jedem Patient gespeichert. 
# Hierzu gehören Name, eine oder mehrere Telefonnummern, eine Adresse (bestehend aus Strasse, PLZ und Stadt) und eine eindeutige Krankenversicherungsnummer.
### BEGIN SOLUTION
g.add_entity("Patient")
g.add_attribute("Patient", "KV_Nummer", is_pk = True)
g.add_attribute("Patient", "Name")
g.add_attribute("Patient", "Telefon", is_multiple = True)
g.add_attribute("Patient", "Adresse", 
                composed_of = ["Straße", "PLZ", "Stadt"])
### END SOLUTION

# 2. Ein Patient kann mehrere Ärzte besuchen. Ein Arzt kann mehrere Patienten behandeln. 
### BEGIN SOLUTION
g.add_relation({"Patient": "m"}, "besucht", {"Arzt": "n"})
### END SOLUTION

# Ein Arzt kann unter anderem Assistenzarzt, Oberarzt oder Chefarzt sein.
### BEGIN SOLUTION
g.add_is_a("Arzt", ["Assistenzarzt", "Oberarzt", "Chefarzt"], is_total = False, is_disjunct = True)
### END SOLUTION

# ---------------------------
# Zwischenergebnis speichern
# ---------------------------
diags.append({"descr": "ER Diagramm 1. - 2.", "diag": g})
g = ER()

# 3. Ein Patient kann ein oder mehrere Medikamente erhalten für eine oder mehrere Behandlungen.
### BEGIN SOLUTION
g.add_relation(
    {"Patient": "1"}, "erhält", {
        "Medikament": "n",
        "Behandlung": "n"
    })
### END SOLUTION

# 4. Jedes Medikament hat eine eindeutige Pharmazentralnummer (PZN).
### BEGIN SOLUTION
g.add_attribute('Medikament', 'Pharmazentralnummer (PZN)', is_pk = True)
### END SOLUTION

# 5. Jede Behandlung hat eine eindeutige ID und Beschreibung.
### BEGIN SOLUTION
g.add_attribute('Behandlung', 'ID', is_pk = True)
g.add_attribute('Behandlung', 'Beschreibung')
### END SOLUTION

# ---------------------------
# Zwischenergebnis speichern
# ---------------------------
diags.append({"descr": "ER Diagramm 3. - 5.", "diag": g})
g = ER()

# 6. Ein Arzt verschreibt ein oder mehrere Medikamente. Ein Medikament kann von einem oder mehreren Ärzten verschrieben werden.
### BEGIN SOLUTION
g.add_relation({"Arzt": "m"}, "verschreibt", {"Medikament": "n"})
### END SOLUTION

# 7. Ein Mitarbeiter ist entweder ein Arzt oder ein Krankenpfleger. Andere Rollen werden nicht erwartet. 
### BEGIN SOLUTION
g.add_is_a("Mitarbeiter", ["Arzt", "Krankenpfleger"], is_total = True, is_disjunct = True)
### END SOLUTION

# 7. Zu jedem Mitarbeiter werden Name, Lebenslauf, eine eindeutige Personalnummer und eine oder mehrere Telefonnummern gespeichert.
# ### BEGIN SOLUTION 
g.add_attribute('Mitarbeiter', 'Personalnummer', is_pk = True)
g.add_attribute('Mitarbeiter', 'Name')
g.add_attribute('Mitarbeiter', 'Telefonnummern', is_multiple = True)
### END SOLUTION

# Ein Mitarbeiter wird von einem Mitarbeiter beaufsichtigt, ein Mitarbeiter beaufsichtigt mehrere Mitarbeiter.
### BEGIN SOLUTION
g.add_relation({"Mitarbeiter": "1"}, "beaufsichtigt", {"Mitarbeiter": "n"})
### END SOLUTION

# ---------------------------
# Zwischenergebnis speichern
# ---------------------------
diags.append({"descr": "ER Diagramm 6. - 7.", "diag": g})
g = ER()

# 8. Ein Krankenpfleger arbeitet auf genau einer Station. In einer Station arbeiten mehrere Krankenpfleger.
### BEGIN SOLUTION
g.add_relation({"Krankenpfleger": "n"}, "arbeitet", {"Station": "1"})
### END SOLUTION

# 9. Eine Station hat eine eindeutige Bezeichnung.
# Eine Station enthält mehrere Zimmer. Die Zimmernummer ist innerhalb einer Station eindeutig.
### BEGIN SOLUTION
g.add_attribute('Station', 'Bezeichnung', is_pk = True)

g.add_entity('Zimmer', is_weak = True)
g.add_relation({"Station": "1"}, "hat", {"Zimmer": { "cardinality": "n", "is_weak": True } })
g.add_attribute('Zimmer', 'Zimmernummer', is_pk = True, is_weak = True)
### END SOLUTION

# 10. Jedem Patienten ist an einem Datum genau ein Zimmer zugeteilt. In jedem Zimmer ist immer genau ein Patient.
### BEGIN SOLUTION
g.add_relation({"Patient": "1"}, "zugeteilt", {"Zimmer": "1"})

g.add_attribute('zugeteilt', 'Datum')
### END SOLUTION

# ---------------------------
# Zwischenergebnis speichern
# ---------------------------
diags.append({"descr": "ER Diagramm 8. - 10.", "diag": g})


# ------------------------------------------------------
#
#  DIESEN TEIL NICHT VERÄNDERN, SONST GIBT ES 0 PUNKTE!
#
# ------------------------------------------------------
# ---------------------------
# ER Diagramme darstellen
# ---------------------------
h = ER()
for i in diags:
    display(i["descr"]) # Beschreibung darstellen
    h = merge_er_diagrams(h, i["diag"]) # Teil-Diagramme zusammenführen
    i["diag"].display() # Teil-Diagramm darstellen

display("ER Diagramm Krankenhaus-Management-System") # Beschreibung darstellen
h.display() # Gesamt-Diagramm darstellen

# Ergebnis speichern
subtask = SubTask( task0, "a", points = 0 )
subtask.setSolution(h.as_solution())
subtask.setData(h)
```
